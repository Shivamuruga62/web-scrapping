## scrapping website

```python
url=('https://www.flipkart.com/search?q=mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=mobiles%7CMobiles&requestId=66082f5a-26d5-44fc-91da-027dccb50e32')
```

## website response

```python
res=requests.get(url)
```

```python
res
```

## parsing

```python
soup=BeautifulSoup(res.text)
```

```python
soup
```

## football names

```python
moblie=soup.select('._4rR01T')
```

```python
moblie
```

```python
price=soup.select('._1_WHN1')
```

```python
price
```

```python
storage=soup.select('.rgWa7D:nth-child(1)')
```

```python
storage
```
